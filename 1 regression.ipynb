```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
matplotlib.rc('font', size=18)
```

```python
houses = pd.read_csv('/Users/vyacheslavzulanov/JupyterNotebook/Skillbox/1/1/1.4_houses.csv')
houses.head()
```

```python
fig = plt.figure(figsize=(10, 10))
ax = plt.axes()

ax.scatter(houses['dim_1'], houses['price'], s=100)

plt.show()
```

price = price(dim_1) = a * dim_1 + b | b * dim_1 + a

```python
from sklearn.linear_model import LinearRegression
```

```python
X = houses[['dim_1']]
y = houses['price']

reg = LinearRegression().fit(X, y)
```

```python
[b] = reg.coef_ 
a = reg.intercept_
```

```python
def reg_prediction (dim_1):
    return a + b * dim_1
```

```python
reg_prediction(X.dim_1[0])
```

```python
reg.predict(X[0:1])[0]
```

```python
fig = plt.figure(figsize=(10, 10))
ax = plt.axes()

ax.scatter(houses['dim_1'], houses['price'], s=100)
ax.plot([X.dim_1.min(), X.dim_1.max()], [reg_prediction(X.dim_1.min()), reg_prediction(X.dim_1.max())], c='red')

plt.show()
```

```python
fig = plt.figure(figsize=(10, 10))
ax = plt.axes(projection='3d')

ax.scatter(houses['dim_1'],houses['dim_2'], houses['price'], s=100)

ax.set_xlabel('dim_1')
ax.set_ylabel('dim_2')
ax.set_zlabel('price')

plt.show()
```

```python
X = houses[['dim_1', 'dim_2']]
y = houses['price']

reg = LinearRegression().fit(X, y)
```

```python
[b1, b2] = reg.coef_ 
a = reg.intercept_
```

```python
def reg_prediction (dim_1, dim_2):
    return a + b1 * dim_1 + b2 * dim_2
```

```python
reg_prediction(X.dim_1[0], X.dim_2[0])
```

```python
reg.predict(X[0:1])[0]
```

```python
d1, d2 = list(), list()
for x in np.linspace(min(houses['dim_1']), max(houses['dim_1']), 100):
    for y in np.linspace(min(houses['dim_2']), max(houses['dim_2']), 100):
        d1.append(x)
        d2.append(y)
d1 = np.array(d1).reshape(-1, 1)
d2 = np.array(d2).reshape(-1, 1)
p = reg.predict(np.concatenate([d1, d1], axis=1))
```

```python
fig = plt.figure(figsize=(20, 20)) 
ax = plt.axes(projection="3d") 


ax.scatter(houses["dim_1"], houses["dim_2"], houses["price"], s=100) 


ax.plot_trisurf(d1.ravel(), d2.ravel(), p.ravel(), alpha=0.2) 



ax.set_xlabel("dim_1") 
ax.set_ylabel("dim_2") 
ax.set_zlabel("price") 



ax.elev = 27 
plt.show()
```
